**The syntax for**
- if-else statements in Python uses indentation to define code blocks, rather than braces or other delimiters. 
- Python also uses the keyword elif (short for "else if") to handle multiple conditions

**Basic if Statement**
- The simplest conditional statement checks a condition and runs a code block only if the condition is True

```python
if condition:
    # Code to execute if the condition is True
    print("The condition is True")
```

- A colon (:) follows the condition.
- The code within the if block is indented (usually four spaces). 

**if-else Statement**
- The if-else statement provides an alternative block of code to run if the initial condition is False

```python
if condition:
    # Code to execute if the condition is True
    print("The condition is True")
else:
    # Code to execute if the condition is False
    print("The condition is False")
```

- The else keyword is followed by a colon (:).
- The else block must have the same indentation level as the if statement it belongs to

**if-elif-else Statement**
- For handling multiple conditions, the if-elif-else structure (also known as an "if ladder") is used. Conditions are checked sequentially, and the first block with a True condition is executed

```python
if condition1:
    # Code to execute if condition1 is True
    print("Condition 1 met")
elif condition2:
    # Code to execute if condition1 is False and condition2 is True
    print("Condition 2 met")
else:
    # Code to execute if none of the above conditions are True
    print("No condition met")
```

- You can have any number of elif blocks.
- Only one block of code will be executed in an if-elif-else chain

**One-Line if-else (Ternary Operator)** 
- Python offers a concise syntax for simple if-else statements, often called the ternary operator, which is useful for assigning a value based on a condition

```python
result = value_if_true if condition else value_if_false
```

- Example:

```python
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)
# Output: Adult
```

```python
# 1. Min of 3 Numbers
a = int(input('first num: '))
b = int(input('second num: '))
c = int(input('third num: '))

if a < b and a < c:
    print('smallest is', a)
elif b < c:
    print('smallest is', b)
else:
    print('smallest is', c)
```
